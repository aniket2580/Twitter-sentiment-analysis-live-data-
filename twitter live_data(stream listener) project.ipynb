```python
#So far we have used our own model, now we will perform sentiment analysis using vader sentiment analyzer
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentimentanalyzer=SentimentIntensityAnalyzer()
```

```python
data['vader analyzer']=data['Text']
```

```python
def vader_analyze(s):
    score=sentimentanalyzer.polarity_scores(s)
    if score['compound'] >= 0.05 : 
        return 1 
  
    elif score['compound'] <= - 0.05 : 
        return -1 
  
    else : 
        return 0
```

```python
data['vader analyzer']=data['vader analyzer'].apply(vader_analyze)
```

```python
ypreds=list(data['vader analyzer'])
```

```python
#making two variables to store the count of positive and negative and neutral tweets predicted by the vader model
positives=0
negatives=0
neutral=0
for x in ypreds:
    if(x==1):
        positives=positives+1
    if(x==-1):
        negatives=negatives+1
    else:
        neutral=neutral+1
```

```python
print('positive percentage for',keyword,' using vader is:',positives/len(ypreds)*100)
print('negative percentage for',keyword,' using vader is:',negatives/len(ypreds)*100)
print('neutral percentage for',keyword,' using vader is:',neutral/len(ypreds)*100)
```
